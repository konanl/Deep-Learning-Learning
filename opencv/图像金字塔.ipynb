```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## 高斯金字塔

```python
img = cv2.imread('./fig/lenna.png')
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
dst = cv2.pyrDown(img_rgb)
```

```python
plt.imshow(dst)
```

```python
print(img_rgb.shape, dst.shape)
```

```python
dst_up = cv2.pyrUp(dst)
```

```python
plt.imshow(cv2.pyrUp(dst_up))
```
